```python
# Define bins and labels for categorization
bins = [-float('inf'), 1.8, 3.0, float('inf')]
labels = ['Distress Zone', 'Grey Zone', 'Safe Zone']

# Categorize companies into risk zones
df['Risk_Zone'] = pd.cut(df['Altman Z-Score'], bins=bins, labels=labels)

# Print the results or save to a new CSV file
print(df[['Ticker', 'Altman Z-Score', 'Risk_Zone']])
```
